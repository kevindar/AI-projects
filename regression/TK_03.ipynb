```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
class gradient:
    def __init__(self, m, b):
        self.m = m
        self.b = b

def linear_regression(gradient, i):
    return gradient.m * (i+1) + gradient.b

def cost_function(trainingSet, gradient):
    sum = 0
    arrLen = len(trainingSet)
    for i in range(arrLen):
        hipotesa = linear_regression(gradient, i)
        real = trainingSet[i]
        sum += (hipotesa - real)**2
    return sum / (2 * arrLen)

def test(trainingSet, m, mStep, mTry, b, bStep, bTry):
    gradientBest = 99
    costBest = 99
    for i in range(mTry):
        for j in range(bTry):
            mTest = m + mStep * i
            bTest = b + bStep * j
            gradientTest = gradient(mTest, bTest)
            result = cost_function(trainingSet, gradientTest)
            if (result < costBest):
                costBest = result
                gradientBest = gradient(mTest, bTest)
                # 0 is the best cost we can get, so return anyway
                if (result == 0): return gradientBest
    return gradientBest

trainingSet = [ 1, 2, 3 ]
gradient = test(trainingSet, 0, 0.5, 100, 0, 0.5, 100)
print(gradient.m)
print(gradient.b)
```
